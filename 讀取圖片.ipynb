```python
import tensorflow as tf
import matplotlib.image as mp_image
import matplotlib.pyplot as plt
```

```python
# raw_img = mp_image.imread("PainImage/059-fn059/fn059t2aiaff/fn059t2aiaff190.png")
raw_img = mp_image.imread("packt.jpeg")
```

```python
print("ndim  = {}".format(raw_img.ndim))
print("shape = {}".format(raw_img.shape))
```

```python
plt.imshow(raw_img)
plt.show()
```

```python
my_image = tf.placeholder(dtype=tf.uint8, shape=(None, None, 3))
my_slice = tf.slice(input_ = my_image, begin=[0, 0, 0], size=[30, 100, 3])
```

```python
with tf.Session() as sess:
    result = sess.run(my_slice, feed_dict={my_image: raw_img})
    print(type(result))
    print(result.shape)
    plt.imshow(result)
    plt.show()
```
