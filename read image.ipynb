## 透過與 matplotlib 的比較，了解 tensorflow 讀寫影像的流程

```python
import tensorflow as tf
import matplotlib.image as mp_image
import matplotlib.pyplot as plt
```

```python
# 預設 Session
sess = tf.Session()

# 預設圖片路徑
img_path = "PainImage/059-fn059/fn059t2aiaff/fn059t2aiaff190.png"
```

```python
ts_img = tf.image.decode_png(tf.read_file(img_path)) # tensorflow 讀取影像
mp_img = mp_image.imread(img_path) # matplotlib 讀取影像
```

```python
print("{}, {}".format(type(ts_img.eval(session=sess)), ts_img.eval(session=sess).shape))
print("{}, {}".format(type(mp_img), mp_img.shape))
```

```python
# 最後一個元素比較
print("{}".format(ts_img.eval(session=sess)[-1][-1]))
print("{}".format(mp_img[-1][-1]))
```

```python
# matplotlib.image 讀取影像呈現
plt.imshow(mp_img)
plt.show()
```

```python
# tf.image 讀取影像呈現
plt.imshow(ts_img.eval(session=sess))
plt.show()
```
